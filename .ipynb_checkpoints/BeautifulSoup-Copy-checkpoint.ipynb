```python
# https://www.dataquest.io/blog/web-scraping-tutorial-python/

page = requests.get("http://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168")
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())

```

```python
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)
```

```python
page = requests.get("https://markets.businessinsider.com/stocks")
print(page.status_code)
soup = BeautifulSoup(page.content, 'html.parser')
stocks = soup.find(id="shares_topflop_StockPricesSharesTopFlop")
item = stocks.find('a')
print(item)

prices = stocks.find_all(class_="row-hover")
for price in prices:
     print(price.get_text())
        
        
print(prices)

names = []
for i in range(len(prices)):
    names.append(prices[i].find('a').get_text())

print(names)
```

```python
# https://docs.python.org/3/library/re.html
import re 

bruh = stocks.find_all('a', text = re.compile("JPMorgan"))
print(bruh)
```

```python
# https://stackoverflow.com/questions/47928608/how-to-use-beautifulsoup-to-parse-google-search-results-in-python
# quick example of using BeautifulSoup to Google for you 

import urllib
from bs4 import BeautifulSoup
import requests
import webbrowser

text = 'hello world'
text = urllib.parse.quote_plus(text)

url = 'https://google.com/search?q=' + text

response = requests.get(url)

with open('output.html', 'wb') as f:
    f.write(response.content)
webbrowser.open('output.html')

soup = BeautifulSoup(response.text, 'html.parser')

for g in soup.find_all(class_='BNeawe vvjwJb AP7Wnd'):
    print(g.get_text())
    print('-----')
```

```python
# Prints out the top 3 videos on YouTube's trending page (aka hot garbage simulator)

import requests
from bs4 import BeautifulSoup

page = requests.get("https://www.youtube.com/feed/trending")

soup = BeautifulSoup(page.content, 'html.parser')
prettysoup = soup.prettify()

videos = list(soup.find_all(class_="yt-lockup-content"))

for i in range(3):
    ele = videos[i]
    
    title = ele.find(class_="yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink spf-link").get_text()
    channel = ele.find(class_="yt-uix-sessionlink spf-link").get_text()
    duration = ele.find(class_="accessible-description").get_text()
    views = ele.find(class_="yt-lockup-meta-info").find_all("li")[1].get_text()
    description = ele.find(class_="yt-lockup-description yt-ui-ellipsis yt-ui-ellipsis-2").get_text()
    time = ele.find(class_="yt-lockup-meta-info").find_all("li")[0].get_text()
    
    print(title)
    print(channel)
    print(duration[3:-1])
    print(views)
    print(description)
    print(time)
    print("")
```
